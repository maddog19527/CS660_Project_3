#EDA on Digits Dataset

```python
# Import libraries
import numpy as np
import pandas as pd

# Load digits dataset
from sklearn import datasets
digits = datasets.load_digits()

# Convert to DataFrame for easier manipulation
df_digits = pd.DataFrame(digits.data, columns=[f'Pixel_{i}' for i in range(digits.data.shape[1])])
df_digits['Target'] = digits.target  # Add the target column

# Display basic information
print("Shape of the dataset:", df_digits.shape)
```

```python
df_digits.head()
```

```python
df_digits.info()
```

```python
df_digits.describe()
```

```python
# Check for missing values
missing_values = df_digits.isnull().sum()
print("\nMissing values in each column:\n", missing_values[missing_values > 0])

# Verify no missing values
if missing_values.sum() == 0:
    print("\nNo missing values in the dataset!")
```

###Distribution of Target Labels

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the frequency of each digit
target_counts = df_digits['Target'].value_counts()

# Plot the distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=target_counts.index, y=target_counts.values, palette='viridis')
plt.title("Distribution of Target Digits")
plt.xlabel("Digits")
plt.ylabel("Frequency")
plt.show()
```

###Correlation Heatmap

```python
# Compute correlation matrix
correlation_matrix = df_digits.corr()

# Plot heatmap for the first 20 pixels to avoid overcrowding
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix.iloc[:20, :20], annot=False, cmap='coolwarm', cbar=True, vmin=-1, vmax=1)
plt.title("Correlation Heatmap (First 20 Pixels)")
plt.show()
```

###Feature Variance

```python
# Calculate variance for each feature
feature_variances = df_digits.var().drop('Target')

# Plot feature variances
plt.figure(figsize=(10, 5))
sns.lineplot(x=range(len(feature_variances)), y=feature_variances, color='blue')
plt.title("Feature Variances")
plt.xlabel("Feature Index")
plt.ylabel("Variance")
plt.show()

# Identify low-variance features
low_variance_features = feature_variances[feature_variances < 0.1]
print("\nLow-variance features:\n", low_variance_features)
```

###PCA Visualization (2D Projection)

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce dimensions to 2
pca = PCA(n_components=2)
digits_pca = pca.fit_transform(digits.data)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=digits_pca[:, 0], y=digits_pca[:, 1], hue=digits.target, palette='tab10', s=60)
plt.title("2D PCA Projection of Digits Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Digits", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

###Feature Distributions

```python
# Filter data for digit '0'
digit_zero = df_digits[df_digits['Target'] == 0]

# Plot distributions of first 5 pixel values
plt.figure(figsize=(12, 6))
for i in range(5):
    sns.kdeplot(digit_zero[f'Pixel_{i}'], label=f'Pixel_{i}', fill=True)
plt.title("Distribution of Pixel Values for Digit '0'")
plt.xlabel("Pixel Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

###Pairplot of Selected Pixels

```python
# Select pixels and create pairplot
selected_pixels = ['Pixel_0', 'Pixel_10', 'Pixel_20', 'Pixel_30', 'Target']
sns.pairplot(df_digits[selected_pixels], hue='Target', palette='tab10')
plt.suptitle("Pairplot of Selected Pixels", y=1.02, fontsize=14)
plt.show()
```

### Pixel Intensity Heatmap

```python
# Compute mean pixel intensity for each digit
mean_pixel_intensity = df_digits.groupby('Target').mean().iloc[:, :64]

# Reshape and plot as heatmaps
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
for i, ax in enumerate(axes.ravel()):
    digit_heatmap = mean_pixel_intensity.iloc[i].values.reshape(8, 8)
    sns.heatmap(digit_heatmap, ax=ax, cmap='viridis', cbar=False, annot=False)
    ax.set_title(f"Digit: {i}")
    ax.axis('off')
plt.title("Average Pixel Intensity for Each Digit", fontsize=16)
plt.show()
```

#Task 2: Classification on the Digits Dataset

```python
# Import necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the digits dataset
from sklearn.datasets import load_digits
digits = load_digits()
X_digits = digits.data  # Features
y_digits = digits.target  # Labels
```

```python
# Step 1: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits, test_size=0.3, random_state=42)
```

```python
# Step 2: Train a DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(random_state=42)  # Initialize Decision Tree
dt_classifier.fit(X_train, y_train)  # Train on training data
```

```python
# Step 3: Train a RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=42)  # Initialize Random Forest
rf_classifier.fit(X_train, y_train)  # Train on training data
```

```python
# Step 4: Make predictions on the test set
y_pred_dt = dt_classifier.predict(X_test)  # Predictions from Decision Tree
y_pred_rf = rf_classifier.predict(X_test)  # Predictions from Random Forest
```

```python
# Step 5: Evaluate the models using confusion matrices
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
```

```python
# Plot confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.heatmap(conf_matrix_dt, annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Decision Tree Confusion Matrix")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="Blues", ax=axes[1])
axes[1].set_title("Random Forest Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()
```

```python
# Step 6: Print classification reports for both models
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
```

#Task 3 Tuning Classifiers

**Hyperparameter Tuning:**
- Use GridSearchCV to tune parameters for:
 - DecisionTreeClassifier: Depth and splitting criteria.
 - RandomForestClassifier: Number of trees and max depth.
- Train with Optimal Parameters:
  - Use the best parameters from GridSearchCV to retrain both models.
- Evaluate Models:
  - Compute confusion matrices and evaluate performance.
- Comparison:
  - Determine which model performs better post-tuning.

### **Step 1: Define Parameter Grids for Both Models**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Parameter grid for DecisionTreeClassifier
dt_param_grid = {
    'max_depth': [5, 10, 15, None],
    'criterion': ['gini', 'entropy']
}

# Parameter grid for RandomForestClassifier
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'criterion': ['gini', 'entropy']
}
```

### **Step 2: Tune Parameters Using GridSearchCV**
###**And**
### **Step 3: Train with Optimal Parameters**

```python
# Perform GridSearchCV for DecisionTreeClassifier
dt_grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42), dt_param_grid, cv=5, scoring='accuracy')
dt_grid_search.fit(X_train, y_train)
best_dt_model = dt_grid_search.best_estimator_  # Best Decision Tree model

# Perform GridSearchCV for RandomForestClassifier
rf_grid_search = GridSearchCV(RandomForestClassifier(random_state=42), rf_param_grid, cv=5, scoring='accuracy')
rf_grid_search.fit(X_train, y_train)
best_rf_model = rf_grid_search.best_estimator_  # Best Random Forest model
```

### **Lets Evaluate**

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Make predictions using the tuned models
y_pred_dt_tuned = best_dt_model.predict(X_test)
y_pred_rf_tuned = best_rf_model.predict(X_test)

# Compute confusion matrices
conf_matrix_dt_tuned = confusion_matrix(y_test, y_pred_dt_tuned)
conf_matrix_rf_tuned = confusion_matrix(y_test, y_pred_rf_tuned)

# Plot confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.heatmap(conf_matrix_dt_tuned, annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Tuned Decision Tree Confusion Matrix")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

sns.heatmap(conf_matrix_rf_tuned, annot=True, fmt="d", cmap="Blues", ax=axes[1])
axes[1].set_title("Tuned Random Forest Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()

# Print classification reports for both models
print("Tuned Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt_tuned))
print("Tuned Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf_tuned))
```

### **Comparison: Original vs. Tuned Models**

#### **Decision Tree: Original vs. Tuned**
- **Accuracy Improvement**: Increased from **84%** to **88%** after tuning.
- **Model Performance**: Precision and recall improved for most digits, especially `1`, `2`, and `3`. However, it still struggles with digits like `4` and `8`.
- **Conclusion**: Tuning made the model better but didn't eliminate inconsistencies. It still has limited generalization ability.

#### **Random Forest: Original vs. Tuned**
- **Accuracy Consistency**: Stayed at **98%** before and after tuning.
- **Model Performance**: Both versions provided near-perfect precision, recall, and F1-scores across all classes, maintaining robustness.
- **Conclusion**: Tuning had little effect since the model was already well-optimized. The Random Forest consistently delivers high accuracy and reliable predictions.

### **Overall Summary**
- **Decision Tree** saw a modest improvement with tuning but still lacks robustness.
- **Random Forest** remains the best model, maintaining an accuracy of **98%** with consistent performance across all classes.

**Preferred Model**: The **Random Forest** is preferred due to its high reliability and strong generalization, both with and without tuning.
